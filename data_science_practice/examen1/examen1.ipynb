```python
import pandas as pd
import numpy as np
import re
import os

file_path = 'pakistan_intellectual_capital.csv'
df = pd.read_csv(file_path)
df
```

## Step 1

Primero revisaremos la columna “Country” para asegurarnos que no hay inconsistencias. Para esto prepara un arreglo separado denominado “Countries” en donde guardemos los valores únicos de esa columna. Después ordénalos alfabéticamente y presenta en pantalla el resultado.


```python
# Check all variations of country
Countries = df['Country'].sort_values().apply(lambda x: x.strip().capitalize()).unique()

# Show findings
print(f"Amount of countries: {Countries.size}")
Countries
```


## Step 2

Revisa si en este arreglo hay inconsistencias y resuelve como eliminarlas en el DataFrame original.

```python
# Apply formatting to all countries
df['Country'] = df['Country'].apply(lambda x: x.strip().capitalize())

# Replase USofA with USA
df['Country'] = df['Country'].replace('Usofa', 'USA')
df['Country'] = df['Country'].replace('Usa', 'USA')
df['Country'] = df['Country'].replace('Uk', 'UK')

print(f"Amount of countries: {df['Country'].unique().size}")
df['Country'].value_counts().sort_index(ascending=True)
```

## Step 3

Revisemos ahora inconsistencias en la columna “Graduated from”, revisando los valores únicos y resolviendo detalles en el DataFrame

```python
graduated = df['Graduated from'].unique()
graduated.sort()
print(graduated)

#Elimiar primero por comas
df['Graduated from'] = df['Graduated from'].str.split(',').str[0]
graduated = df['Graduated from'].unique()
graduated.sort()
print(graduated)

df['Graduated from'] = df['Graduated from'].str.strip().str.replace(r'[^\w\s]', '', regex=True)
graduated = df['Graduated from'].unique()
graduated.sort()
print(graduated)

#Elimanr primero caracteres especiales
df ['Graduated from'] = df['Graduated from'].str.extract('([a-zA-Z\s]+)', expand=False).str.strip()
graduated = df['Graduated from'].unique()
graduated.sort()
for i in graduated:
  print(i)

  #Funcion para estandarisas mayusculas y of

mantener = {"of","and","for"}
df['Graduated from'] = df['Graduated from'].apply(lambda text: ' '.join([word.lower() if word.lower() in mantener and i > 0 else word.capitalize() for i, word in enumerate(text.split())]))
graduated = df['Graduated from'].unique()
graduated.sort()
for i in graduated:
    print(i)
```

## Step 4

El “Terminal Degree” solo puede tener cuatro categorías: “Bacherlor”, “Master”, “Doctorate” y “PostDoctorate”. De acuerdo a esto realiza las correcciones de inconsistencias en la columna.

```python
# Define HashMap for all Terminal Degrees
terminal_degree = {'b': 'Bacherlor', 
                   'm': 'Master', 
                   'ph': 'Doctorate', 
                   'd': 'Doctorate',
                   'po': 'PostDoctorate'
                   }

# Verify all degrees
df['Terminal Degree'].sort_values().apply(lambda x: str(x).strip().lower()).unique()
```

```python
# Check how many records are NaN
records_nan = df['Terminal Degree'].isna().sum()
records_nan
```

```python
# Drop all NaN records since we can't determine the degree, it's only 4 records out of 251 and we want the preferences per degree
df = df.dropna(subset=['Terminal Degree'])


# Apply formatting to all degrees
for key, value in terminal_degree.items():
    df['Terminal Degree'] = df['Terminal Degree'].apply(lambda x: value if str(x).strip().lower().startswith(key) else x)

# Verify all degrees
df['Terminal Degree'].value_counts().sort_index(ascending=True)
```

## Step 5

El campo “Area of Specialization / Research Interests” contiene de manera opcional varias areas de estudio en un solo registro. Divide estas en varias columnas denominadas “Area of Interest 1”, “Area of Interest 2” y así sucesivamente según el número de áreas que tenga un alumno. Para esto define los separadores comunes y utilízalos para cortar los valores.

```python
# Format column to only contain a list of areas of interest
# So to create the regex, we need to find the most common separator which are [',', 'and', '&', '/']
areas_of_interest = df['Area of Specialization/Research Interests'].str.split(r'/|&|,|;|and', expand=True)
areas_of_interest = areas_of_interest.apply(lambda x: x.str.strip(), axis=1)
areas_of_interest = areas_of_interest.apply(lambda x: x.str.lower(), axis=1)
```

```python
areas_of_interest.sample(4)
```

```python
# Create a list of all areas of interest
areas_of_interest.columns = ['Area of Interest {}'.format(n) for n in range(1, len(areas_of_interest.columns) + 1)]
areas_of_interest

# Append to the end of the dataframe
df = pd.concat([df, areas_of_interest], axis=1)
df.sample(4)
```

## Step 6

Por último, necesitamos saber para nuestro análisis el top 3 de áreas de interés más populares por cada grado de estudios.

```python
# Create counters for each area of interest with stack
areas_of_interest = areas_of_interest.stack().value_counts()
areas_of_interest.dropna(inplace=True)
```

```python
# Print the top 3 areas of interest
print(f"Top 3 areas of interest (area: amount_people_interested):")

for i in range(3):
    print(f"{i+1}. {areas_of_interest.index[i]}: {areas_of_interest.iloc[i]}")
```
