```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos de una distribución uniforme
data_uniforme = np.random.uniform(0, 1, 1000)

# Crear histograma
plt.hist(data_uniforme, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma de Distribución Uniforme')
plt.show()
```

```python
import seaborn as sns
# Crear boxplot
sns.boxplot(data=data_uniforme, color='lightblue')
plt.title('Boxplot de Distribución Uniforme')
plt.show()
```

```python
import numpy as np

# Generar un dataset aleatorio con 1000 observaciones
np.random.seed(42)
#Genera 970 datos siguiendo una distribución normal (gaussiana) con una media (loc) de 50 y una desviación estándar (scale) de 10.
data = np.random.normal(loc=50, scale=10, size=970)  # 97% de los datos

#Genera 30 datos siguiendo una distribución normal con una media de 150 y una desviación estándar de 20.
#Estos 30 datos representan el 3% del total y son considerados outliers debido a su media significativamente diferente y mayor dispersión.
outliers = np.random.normal(loc=150, scale=20, size=30)  # 3% de outliers
data = np.concatenate((data, outliers))

# Mezclar los datos
np.random.shuffle(data)
```

```python
# Formatear el número con 2 decimales
# Calcular la media, moda, mediana y desviación estándar
mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)
cv=std_dev/mean
min=np.min(data)
max=np.max(data)
perc25 = np.percentile(data,25)
perc75 = np.percentile(data,75)
inter=perc75-perc25

print(f"Media: {mean:.2f}")
print(f"Mediana: {median:.2f}")
print(f"Desviación Estándar: {std_dev:.2f}")
print(f"Coeficiente de variacion: {cv:.2f}")
print(f"Percentil 25:{perc25:.2f}")
print(f"Percentil 75:{perc75:.2f}")
print(f"Intercuartil: {inter:.2f}")
print(f"Valor mínimo: {min:.2f}")
print(f"valor máximo: {max:.2f}")
```

```python
# Crear histograma
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma del Dataset Aleatorio con Outliers')
plt.show()

# Crear boxplot
sns.boxplot(data=data, color='lightblue')
plt.title('Boxplot del Dataset Aleatorio con Outliers')
plt.show()
```
