```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
label = LabelEncoder()
lb = label.fit_transform(df["sex"])
Data = df.drop("sex",axis='columns')
Data["sex"] = lb
```

```python
Data.head()
```

```python
XData = Data[['total_bill','tip','sex']]
```

```python
XData.head()
```

```python
scaler = MinMaxScaler()
XData[['total_bill']] = scaler.fit_transform(XData[['total_bill']])
XData
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
ScaledData = scaler.fit_transform(XData)
```
