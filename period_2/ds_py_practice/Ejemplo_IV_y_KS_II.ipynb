```python
# Función para generar el cuadro de frecuencia y calcular IV y KS por cada grupo
def calculate_iv_ks_table_by_groups(data, feature, target):
    # Generar el cuadro de frecuencia con IV y KS por grupo
    grouped = data.groupby(feature).agg(
        num_observaciones=(target, 'count'),
        tasa_morosidad=(target, 'mean')
    )

    # Calcular el número total de buenos (Malo = 0) y malos (Malo = 1)
    total_good = len(data[data[target] == 0])
    total_bad = len(data[data[target] == 1])

    # Calcular la cantidad de buenos y malos por grupo
    grouped['buenos'] = grouped['num_observaciones'] * (1 - grouped['tasa_morosidad'])
    grouped['malos'] = grouped['num_observaciones'] * grouped['tasa_morosidad']

    # Calcular las proporciones de buenos y malos por grupo
    grouped['% buenos'] = grouped['buenos'] / total_good
    grouped['% malos'] = grouped['malos'] / total_bad

    # Calcular el Information Value (IV) por grupo
    grouped['IV'] = (grouped['% buenos'] - grouped['% malos']) * np.log(grouped['% buenos'] / grouped['% malos'])

    # Calcular la distribución acumulada de buenos y malos para KS
    grouped['acum % buenos'] = grouped['% buenos'].cumsum() * 100  # Convertir a porcentaje
    grouped['acum % malos'] = grouped['% malos'].cumsum() * 100    # Convertir a porcentaje

    # Calcular el KS por grupo
    grouped['KS'] = abs(grouped['acum % buenos'] - grouped['acum % malos'])

    # Calcular el % de información por grupo
    total_observations = len(data)
    grouped['% poblacional'] = (grouped['num_observaciones'] / total_observations) * 100

    # Calcular el IV total sumando los IV de cada grupo
    iv_total = grouped['IV'].sum()

    # Calcular el KS máximo
    ks_max = grouped['KS'].max()

    return grouped, iv_total, ks_max

# Ejecutar la función para calcular los indicadores completos y generar el cuadro de frecuencia
group_summary_final, iv_total_final, ks_total_final = calculate_iv_ks_table_by_groups(disciplina_tech_grouped_final, 'Disciplina_tech_group', 'Malo')

# Mostrar el cuadro de frecuencia con los resultados y IV y KS totales
print(group_summary_final)

# Mostrar IV y KS totales
print(f"IV total: {iv_total_final}, KS total: {ks_total_final}")
```
