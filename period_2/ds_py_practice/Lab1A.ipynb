```python
# 7) Transform the following columns to the specified data types
df['customer_number'] = df['customer_number'].astype('int')
df['customer_name'] = df['customer_name'].astype('string')
df['2016'] = df['2016'].apply(lambda x: x.replace(',', '').replace('$', '')).astype('float')
df['2016'] = df['2016'].astype('float')
df['2017'] = df['2017'].apply(lambda x: x.replace(',', '').replace('$', '')).astype('float')
df['2017'] = df['2017'].astype('float')
df['percent_growth'] = df['percent_growth'].apply(lambda x: x.replace('%', '')).astype('float')
df['percent_growth'] = df['percent_growth'].astype('float')
df['jan_units'] = df['jan_units'].apply(lambda x: x if x.isnumeric() else 0).astype('float')
df['jan_units'] = df['jan_units'].astype('float')
df['active'] = df['active'].replace('Y', True).replace('N', False)
df['active'] = df['active'].astype('bool')

# Display the data types of each column
df.info()
```

```python
# 8) Add a column “Start_date” to the data frame. This column takes de columns Month, Day, Year and generates an array of type “datetime64”
df['start_date'] = pd.to_datetime(df[['month', 'day', 'year']])
# 9) Display the first 3 rows and all columns.
df.head(3)
```
