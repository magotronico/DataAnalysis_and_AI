7. Generate a graph, showing on the “X” axis the density variable, and on the “Y” axis the pH variable. Be sure to use a transparency value less than 0.2. Add plt.xlabel('Density'), plt.ylabel('pH'), plt.grid() and plt.title('Density vs pH scatter plot')
