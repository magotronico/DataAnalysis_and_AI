As a data scientist, you have to ask yourself which is better and make a decision.

Now it is time to move on to the topic of correlations, because it will be important for the linear regression. I will leave the lines of code below, but I will explain what is happening.

-In the label coder cell, what we are going to do is convert the categorical variables that have only two options, gender and income, to 0 and 1's so that we can work with them with the numeric variables.

-In the scoring matrix cell, we want to see if there are any variables that have a linear relationship with each other. Different measures are used, such as Pearson, Spearman, Kendall, etc. In this example, Pearson is used by default. If there is an association of 1, it is perfect and has a positive slope, -1 is the same but with a negative slope, and 0 means that there is no correlation between the variables.

-The last cell is a scatterplot matrix between all these variables.

```python
from sklearn.preprocessing import LabelEncoder
#Create a label encoder object
le = LabelEncoder()
# Label Encoding will be used for columns with 2 or less unique values
le_count = 0
for col in df.columns[0:]:
    if df[col].dtype == 'object':
        if len(list(df[col].unique())) <= 2:
            le.fit(df[col])
            df[col] = le.transform(df[col])
            le_count += 1
print('{} columns were label encoded.'.format(le_count))
```

```python
x_num = df.select_dtypes(include=['int64', 'float64'])
corrmat = x_num.corr()

#Correlation matrix
plt.figure(figsize=(10,10))
sns.set(style="darkgrid")

ax = sns.heatmap(corrmat,
                 vmin=-1, vmax=1, center=0, annot=True,
                 cmap=sns.diverging_palette(20, 220, n=200),
                 square=True)

ax.set_xticklabels(ax.get_xticklabels(),
                 rotation=45,
                 horizontalalignment='right')
```
