```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Violin plot
Dataset: "Tips" Seaborn

```python
# Load the tips dataset
tips = sns.load_dataset('tips')

# Set up the plot
plt.figure(figsize=(18, 8))
sns.set(style="darkgrid")
sns.color_palette(palette='Pastel1')

# Create the violin plot
sns.violinplot(
    x='day',
    y='total_bill',
    hue='sex',
    data=tips,
    split=True,
    inner='box'
)

# Calculate and plot the mean total bill
mean_total_bill = tips['total_bill'].mean()
plt.axhline(mean_total_bill, color='red', linestyle='--')
plt.text(3.2, mean_total_bill + 1, f'Promedio: {mean_total_bill:.2f}', color='red')

# Set the plot labels and title
plt.title('Distribución de Ingresos por Día y Género')
plt.xlabel('Día de la Semana')
plt.ylabel('Total de la Factura')
plt.legend(title="Sex",bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

# Show the plot
plt.show()
```

## Scatter plot
Dataset: 'diamonds' Seaborn

```python
diamonds = sns.load_dataset('diamonds')

# Set up the plot
plt.figure(figsize=(18, 8))
sns.set(style="whitegrid")

# Create the scatter plot
sns.scatterplot(
    x='carat',
    y='price',
    hue='cut',
    data=diamonds,
    palette='PuBuGn'
)

# Set the plot labels and title
plt.title('Relación entre Precio y Carat para diferentes niveles de calidad')
plt.xlabel('Carat')
plt.ylabel('Precio')
plt.legend(title="Calidad del Corte",bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

# Show the plot
plt.show()
```

## Heatmap plot
Dataset: 'flights' Seaborn

```python
flights = sns.load_dataset('flights')

# Set up the plot
plt.figure(figsize=(12, 8))
sns.set(style="white")

# Create the heatmap
flights = flights.pivot(index = 'year', columns = 'month', values = 'passengers')
sns.heatmap(flights, annot=True, fmt='d', cmap='RdBu', cbar_kws={'label': 'Pasajeros'})

# Set the plot labels and title
plt.title('Tendencia de Pasajeros por Mes y Año')
plt.xlabel('Mes')
plt.ylabel('Año')

# Show the plot
plt.show()
```

## Boxplots
Datasets: "tips" Seaborn

```python
np.random.seed(10)
tips["age_group"] = np.random.choice(["Joven", "Adulto", "Senior"], size=len(tips))

# Set up the plot
plt.figure(figsize=(12, 8), dpi=1000)
sns.set(style="whitegrid")

# Create the box plot
sns.boxplot(
    x='age_group',
    y='tip',
    hue='sex',
    data=tips,
    palette='Blues'
)

# Overlay the strip plot
sns.stripplot(
    x='age_group',
    y='tip',
    hue='sex',
    data=tips,
    dodge=True,
    palette='dark:k',
)

# Set the plot labels and title
plt.title('Distribución de Propinas por Grupo Etario y Género')
plt.xlabel('Grupo Etario')
plt.ylabel('Propina (USD)')
plt.legend(title="Género",bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

# Show the plot
plt.show()
```

## Scatter plot
Dataset: "diamonds" Seaborn

```python
# Set up the plot
plt.figure(figsize=(12, 8))
sns.set(style="white")

# Create the scatter plot
sns.scatterplot(
    x='carat',
    y='price',
    hue='cut',
    size='depth',
    data=diamonds,
    palette='viridis',
    sizes=(20, 200),
    edgecolor='gray',
    alpha=0.6
)

# Set the plot labels and title
plt.title('Relación entre Precio, Kilates y Claridad de los Diamantes')
plt.xlabel('Kilates')
plt.ylabel('Precio (USD)')
plt.legend(title="Calidad del Corte",bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

# Show the plot
plt.show()
```
